```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import fractal
```

### Setup the parallel computing clients

```python
from IPython import parallel
clients = parallel.Client()
clients = parallel.Client(profile='default')
print("Number of clients: {0}".format(len(clients)))
```

```python
lview = clients.load_balanced_view()
lview.block = True
dview = clients.direct_view()
```

### Configure the desired parameter settings

```python
Nx = 300
Ny = 200
xmin = -2
xmax = 1
ymin = -1
ymax = 1
max_iter = 20
num_chunks = len(clients)
```

```python
#params = dict(Nx=Nx, Ny=Ny, xmin=xmin, xmax=xmax, ymin=ymin, ymax=ymax, max_iter=max_iter, num_chunks=num_chunks)
#dview.push(params);
```

Only 'a' will be an iterable, and the other parameters will remain the same, so in order to use map, we need to wrap the compute_region function in a one-parameter lambda function:

```python
compute_region_lambda = lambda chunk: fractal.compute_region(chunk, num_chunks, Nx, Ny, xmin, xmax, ymin, ymax, max_iter)
```

### Comparison of results

#### Test using one core, with the local **map** function

Check the execution time:

```python
%timeit map(compute_region_lambda, range(num_chunks))
```

Display the result:

```python
regions_local = map(compute_region_lambda, range(num_chunks))
```

```python
plt.imshow(np.hstack(regions_local), cmap=plt.cm.jet)
```

#### Test using multiple cores, with the IPython Parallel **map** function:

Check the execution time:

```python
%timeit lview.map(compute_region_lambda, range(num_chunks))
```

Display the result:

```python
regions_distributed = lview.map(compute_region_lambda, range(num_chunks))
```

```python
plt.imshow(np.hstack(regions_distributed), cmap=plt.cm.jet)
```
